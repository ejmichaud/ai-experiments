```python
style.use('seaborn-notebook')
# outer = gridspec.GridSpec(4, 2, wspace=0.2, hspace=0.2)

for i, init in islice(enumerate(graph_data), 0, 10):
    mean, std = init
    num_batches_data, eis_data, losses_data, accuracies_data = graph_data[init]
    train_losses, test_losses = zip(*losses_data)
    fig = plt.figure(figsize=(12, 8))
    plt.suptitle(f"N({mean}, {std})")
    
    plt.subplot(3, 1, 1)
    plt.plot(num_batches_data, accuracies_data, color='orange')
    plt.title("Accuracy")
    
    plt.subplot(3, 1, 2)
    plt.plot(num_batches_data, test_losses, color='red')
    yavg = np.mean(test_losses)
    plt.ylim(yavg*0.7, yavg*1.4)
    plt.title("Test Loss")
    
    plt.subplot(3, 1, 3)
    plt.plot(num_batches_data, eis_data, color='blue')
    plt.title("Effective Information")

    plt.subplots_adjust(hspace=0.4)
    
    plt.savefig("plots/different-inits2/N-{}-{}.png".format(mean, std), dpi=300)
    
#     inner = gridspec.GridSpecFromSubplotSpec(3, 1,
#                     subplot_spec=outer[i], wspace=0.2, hspace=0.2)
    
#     plt.title(init)
    
#     ax = plt.Subplot(fig, inner[0])
#     plt.plot(num_batches_data, accuracies_data, color='orange')
#     fig.add_subplot(ax)
    
#     ax = plt.Subplot(fig, inner[1])
#     plt.plot(num_batches_data, test_losses, color='red')
#     yavg = np.mean(test_losses)
#     plt.ylim(yavg*0.5, yavg*1.4)
#     fig.add_subplot(ax)
    
#     ax = plt.Subplot(fig, inner[2])
#     plt.plot(num_batches_data, eis_data, color='blue')
#     fig.add_subplot(ax)
    
```
